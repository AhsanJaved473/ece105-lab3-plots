```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.
```

```python
import numpy as np

# Generate simulated sensor data
seed = 3297  # Last 4 digits of Drexel ID for reproducibility
rng = np.random.default_rng(seed)
n = 200

# Sensor A: mean 25°C, std 3°C, 200 readings (float64)
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n).astype(np.float64)

# Sensor B: mean 27°C, std 4.5°C, 200 readings (float64)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n).astype(np.float64)

# Timestamps: uniformly spaced from 0 to 10 seconds, 200 points (float64)
timestamps = np.linspace(0.0, 10.0, n, dtype=np.float64)

# Verify shapes and dtypes
assert sensor_a.shape == (200,) and sensor_a.dtype == np.float64
assert sensor_b.shape == (200,) and sensor_b.dtype == np.float64
assert timestamps.shape == (200,) and timestamps.dtype == np.float64
print(f"Data generated successfully: {n} samples")
```

```python
import matplotlib.pyplot as plt

# Create scatter plot of sensor data
plt.figure(figsize=(10, 6))
plt.scatter(timestamps, sensor_a, alpha=0.6, label='Sensor A', s=30)
plt.scatter(timestamps, sensor_b, alpha=0.6, label='Sensor B', s=30)
plt.xlabel('Time (seconds)')
plt.ylabel('Temperature (°C)')
plt.title('Sensor Temperature Readings Over Time')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Create histogram of sensor readings
plt.figure(figsize=(10, 6))
plt.hist(sensor_a, bins=20, color='blue', alpha=0.6, label='Sensor A')
plt.hist(sensor_b, bins=20, color='orange', alpha=0.6, label='Sensor B')
plt.xlabel('Temperature (°C)')
plt.ylabel('Frequency')
plt.title('Distribution of Sensor Temperature Readings')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Create box plot comparing sensor readings
plt.figure(figsize=(8, 6))
plt.boxplot([sensor_a, sensor_b], labels=['Sensor A', 'Sensor B'])
plt.ylabel('Temperature (°C)')
plt.title('Box Plot of Sensor Temperature Readings')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
